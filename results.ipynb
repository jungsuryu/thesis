## Graph Results

### All Results (w/ MF as base)

```python
# threshold 
thresholds = [i for i in range(1, 51)]
baseline = 0.8433752
```

```python
# Plot settings
plt.figure(figsize=(15, 7))

plt.subplot(1, 2, 1)
plt.axhline(baseline, 0, 20, color='gray', linestyle='--', label='baseline')
plt.plot(thresholds, c_white.values(), color='tab:red', label='MF (White Sheep)', linewidth=1.3)
plt.plot(thresholds, c_bs.values(), color='tab:blue', linestyle=':', label='Bestseller (Gray Sheep)', linewidth=1.7)
plt.plot(thresholds, c_gsbs.values(), color='tab:orange', linestyle=':', label='GSU Bestseller (Gray Sheep)', linewidth=1.7)
plt.plot(thresholds, c_gsmf.values(), color='tab:green', linestyle=':', label='MF (Gray Sheep)', linewidth=1.7)
plt.plot(thresholds, c_neumf.values(), color='tab:purple', linestyle=':', label='NeuMF (Gray Sheep)', linewidth=1.7)
plt.title('MovieLens: All Individual Results (Cosine)', fontdict={'fontsize':12})
plt.yticks(np.linspace(0.83, 1.14, num=8))
plt.xticks([1]+list(range(5, 51, 5)))
plt.xlabel('Similarity Threshold (=Proportion of Gray Sheep Users)', fontdict={'fontsize':12})
plt.ylabel('RMSE', fontdict={'fontsize':12})
plt.legend(fontsize=10)

plt.subplot(1, 2, 2)
plt.axhline(baseline, 0, 20, color='gray', linestyle='--', label='baseline')
# plt.plot(thresholds, white.values(), color='red', label='White MF')
plt.plot(thresholds, c_wbs.values(), color='tab:blue', label='Weighted Bestseller', linewidth=1.3)
plt.plot(thresholds, c_wgsbs.values(), color='tab:orange', label='Weighted GSU Bestseller', linewidth=1.3)
plt.plot(thresholds, c_wgsmf.values(), color='tab:green', label='Weighted MF', linewidth=1.3)
plt.plot(thresholds, c_wneumf.values(), color='tab:purple', label='Weighted NeuMF', linewidth=1.3)
plt.title('MovieLens: Weighted Results (Cosine)', fontdict={'fontsize':12})
plt.yticks(np.linspace(0.84, 0.882, num=8))
plt.xticks([1]+list(range(5, 51, 5)))
plt.xlabel('Similarity Threshold (=Proportion of Gray Sheep Users)', fontdict={'fontsize':12})
plt.ylabel('RMSE', fontdict={'fontsize':12})
plt.legend(fontsize=10)
```
